```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load dataset from uploaded Excel file
file_path = "/content/Load Data in kW.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Rename and prepare data
df.columns = ['Load']
df['Load_Yesterday'] = df['Load'].shift(1)
df.dropna(inplace=True)
df.rename(columns={'Load': 'Load_Today'}, inplace=True)

# Extract features and target
X = df[['Load_Yesterday']].values
y = df['Load_Today'].values

# Normalize data
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = (y - np.mean(y)) / np.std(y)

# Split data (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), kernel_initializer=tf.keras.initializers.Constant(1),
                          bias_initializer=tf.keras.initializers.Constant(-1))
])

# Compile model with RMSprop optimizer
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.1, rho=0.9),
              loss='mse', metrics=['mae'])

# Train model
model.fit(X_train, y_train, epochs=100, verbose=1)

# Predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Compute metrics
def compute_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return mse, rmse, mae

train_mse, train_rmse, train_mae = compute_metrics(y_train, y_train_pred)
test_mse, test_rmse, test_mae = compute_metrics(y_test, y_test_pred)

print(f"Training MSE: {train_mse}, RMSE: {train_rmse}, MAE: {train_mae}")
print(f"Testing MSE: {test_mse}, RMSE: {test_rmse}, MAE: {test_mae}")
```
